```python
import re

# Defining Conversion metrics
perch_to_sqft = 272.25  
acre_to_sqft = 43560    

def convert_to_sqft(land_size):
    #Using regular expression to match only numerical values
    matches = re.findall(r'\d+\.\d+', land_size)
    if matches:
        numeric_value = float(matches[0])
        # Check if "acres" or "perches" is in the original value
        if "acres" in land_size:
            return numeric_value * acre_to_sqft
        elif "perches" in land_size:
            return numeric_value * perch_to_sqft
    return None  

# Apply the conversion function to the Land size column
df3 = df3.copy()
df3['Land size'] = df3['Land size'].apply(convert_to_sqft)

# Display the DataFrame with the converted column
df3
```

```python
# Renaming the columns by removing the spacings and adding '_'
df4 = df3.copy()
df4.rename(
    columns={"Land size": "Land_size", "House size": "House_size"},
    inplace=True
)

df4
```

```python
# Outliers - Validating Bedroom to Sqft Ration via domain knowledge
df5 = df4[~(df4.House_size/df4.Beds<300)]
df5
```

```python
# Outliers - Validating Bathrooms to Bedrooms Ratio via domain knowledge
df5 = df5[~(df5.Baths>df5.Beds+1)]
df5
```

```python
df5.describe()
```

```python
# Plotting box plots to check for outliers
numerical_columns = ['Price', 'Baths', 'Land_size','Beds','House_size']

for column in numerical_columns :
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df5[column])
    plt.title(f'Boxplot - {column}')
    plt.show()
```

```python
# Removing Outliers using Quantiles

df5 = df5.copy()
for col in numerical_columns:
    Q1 = df5[col].quantile(0.25)
    Q3 = df5[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df5[col] = df5[(df5[col] >= lower_bound) & (df5[col] <= upper_bound)][col]
df5 = df5.dropna()
df5 = df5.reset_index(drop=True)
df5
```

```python
df6 =df5.copy()

# Select only the numeric columns for normalization
numeric_columns = ['Baths', 'Land_size','Beds','House_size']

# Creating  MinMaxScaler instance for numeric feature varialbe and target variables
scaler = MinMaxScaler()
price_scaler = MinMaxScaler()

# Normalizing the columns
df6[numeric_columns] = scaler.fit_transform(df6[numeric_columns])
df6['Price']=price_scaler.fit_transform(df6[['Price']])
df6
```

```python
# One-Hot Encoding (for nominal categorical variables)
dummies = pd.get_dummies(df6.Location)

df6 = pd.concat([df6,dummies.drop(columns=dummies.columns[0])],axis ='columns')
df6.tail()
```

```python
# Dropping the original Location Columns.
df6 = df6.drop('Location',axis='columns')
df6.head(10)
```

```python
# Dividing the feature and target variables
X = df6.drop('Price',axis='columns')
y = df6.Price
y.head(10)
```

```python
# Splitting the data into train and test sets in an 80-20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Create an SVR model
svr_model = SVR()  

# Fit the model to the scaled training data
svr_model.fit(X_train, y_train)

# Predict on the test data
y_pred = svr_model.predict(X_test)

# Calculate Mean Squared Error, R-squared, Mean Absolute Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
%%time

# Define a grid of hyperparameters to search
param_grid = {
    'kernel': ['linear', 'rbf', 'poly'],
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1],
}

# Create a GridSearchCV object with 5 fold - cross-validation
grid_search = GridSearchCV(estimator=svr_model, param_grid=param_grid, 
                           scoring='neg_mean_squared_error', cv=5)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Use the best model for predictions
best_svr_model = grid_search.best_estimator_
y_pred = best_svr_model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
# Scatter plot for SVR model
plt.scatter(y_test, y_pred)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("SVR - Scatter Plot of True vs. Predicted Values")
plt.show()
```

```python
# Calculate residuals for SVR
residuals = y_test - y_pred

# Create a residual plot for SVR
plt.scatter(y_pred, residuals)
plt.xlabel("Predictions")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='-')
plt.title("SVR - Residual Plot")
plt.show()
```

```python
# Create an XGBoost Regression model
model = xgb.XGBRegressor()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate Mean Squared Error, R-squared, Mean Absolute Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")


r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2:.2f}")


mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
# Function to predict price (testing before deployment)
def price_predict(baths, land_size, beds, house_size, location):
    loc_index = np.where(X.columns == location)[0][0]

    # Create an array to hold the scaled input features
    scaled_features = np.zeros(len(X.columns))

    # Set the non-numerical attributes
    scaled_features[loc_index] = 1  # Set the location index

    # Scale the numerical input values
    scaled_numerical_features = scaler.transform([[baths, land_size, beds, house_size]])

    # Replace the corresponding positions in the scaled_features array with scaled numerical values
    scaled_features[0:4] = scaled_numerical_features

    return best_forest.predict([scaled_features])[0]
```

```python
# Predict the scaled output
scaled_output = price_predict(4,5788,5,3480,' Akuressa,  Matara')

# Inverse transform the scaled output to get the original scale
unscaled_output = price_scaler.inverse_transform([[scaled_output]])

print("Predicted Price:", unscaled_output[0][0])
```
